```python
from nytimesarticle import articleAPI
api = articleAPI('d30f2891a34546b69fe7930b8961b24f')
articles = api.search( q = "avengers", begin_date = 20180329, end_date= 20180401)
print(articles)
```

```python
import unicodedata
ls = []
for article in articles['response']['docs']:
    ls.append(article['web_url'].encode('utf-8'))
   

    
```

```python
from bs4 import BeautifulSoup
import requests
import unicodedata

article_text = ""
for x in ls[:]:
    session = requests.Session()
    req = session.get(x)
    soup = BeautifulSoup(req.text, "lxml")
    paras = soup.find_all('p', class_='story-body-text story-content')
    for p1 in paras:
        article_text = article_text + p1.get_text()

with open("avengers_day_data2.txt", "w") as text_file:
    print(article_text, file=text_file)

```
